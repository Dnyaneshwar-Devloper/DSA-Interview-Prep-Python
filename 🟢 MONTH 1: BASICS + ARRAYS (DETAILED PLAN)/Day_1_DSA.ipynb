# 🧠 1. WHAT IS AN ALGORITHM?

👉 Simple meaning:
An **algorithm** is a **step-by-step process to solve a problem**.

### 📌 Example

Problem: Find sum of numbers

Steps:

1. Start
2. Take numbers
3. Add them
4. Print result

👉 That’s an algorithm!

# ⚡ 2. WHAT IS TIME COMPLEXITY?

👉 It tells:

> “How fast or slow your code runs as input increases”

---

## 📊 Types of Complexity

### ✅ O(1) — Constant Time

```python
x = arr[0]
```

✔ Always same time

---

### ✅ O(n) — Linear Time

```python
for i in arr:
    print(i)
```

✔ Runs n times

---

### ✅ O(n²) — Quadratic Time

```python
for i in arr:
    for j in arr:
        print(i, j)
```

✔ Runs n × n times

---

### ✅ O(log n) — Logarithmic

👉 Happens in Binary Search (you’ll learn later)

---

# 📌 SIMPLE ANALOGY

| Input Size | O(n)     | O(n²)     |
| ---------- | -------- | --------- |
| 10         | 10 steps | 100 steps |
| 100        | 100      | 10,000 😱 |

👉 That’s why optimization matters!

```python
largest = second = -1

for num in arr:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print(second)
```

# ✅ 1. Find Maximum Element

```python
arr = [3, 7, 2, 9, 5]

max_val = arr[0]
for num in arr:
    if num > max_val:
        max_val = num

print(max_val)
```

✔ Output: `9`

---

# ✅ 2. Find Minimum Element

```python
arr = [3, 7, 2, 9, 5]

min_val = arr[0]
for num in arr:
    if num < min_val:
        min_val = num

print(min_val)
```

✔ Output: `2`

---

# ✅ 3. Reverse an Array

```python
arr = [1, 2, 3, 4]

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print(arr)
```

✔ Output: `[4, 3, 2, 1]`

---

# ✅ 4. Sum of Elements

```python
arr = [1, 2, 3, 4]

total = 0
for num in arr:
    total += num

print(total)
```

✔ Output: `10`

---

# ✅ 5. Count Even Numbers

```python
arr = [1, 2, 3, 4, 6, 7]

count = 0
for num in arr:
    if num % 2 == 0:
        count += 1

print(count)
```

✔ Output: `3`

---

# ✅ 6. Find Second Largest Element

```python
arr = [3, 7, 2, 9, 5]

largest = second = float('-inf')

for num in arr:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print(second)
```

✔ Output: `7`

# ✅ 7. Check if Element Exists

```python
arr = [1, 2, 3, 4, 5]
target = 3

found = False

for num in arr:
    if num == target:
        found = True
        break

print(found)
```

✔ Output: `True`

# ✅ 8. Find Index of Element

```python
arr = [10, 20, 30, 40]
target = 30

for i in range(len(arr)):
    if arr[i] == target:
        print(i)
```

✔ Output: `2`

# ✅ 9. Calculate Average

```python
arr = [2, 4, 6, 8]

total = 0
for num in arr:
    total += num

avg = total / len(arr)
print(avg)
```

✔ Output: `5.0`

---

# ✅ 10. Find Largest and Smallest Together

```python
arr = [3, 7, 2, 9, 5]

largest = smallest = arr[0]

for num in arr:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print("Largest:", largest)
print("Smallest:", smallest)
```

✔ Output:

```
Largest: 9
Smallest: 2
```
